```python
from IPython.display import HTML
import folium
import pandas as pd
import folium
from IPython.display import HTML
```

```python
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))

def embed_map(map, path="map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```

```python
#load centres CRF
# CRF: point rouges
centres = pd.read_excel("../data/croixrouge/structures_2015_CRF.xlsx")

# Draw map
map = folium.Map(location=[50.25, 3.64], zoom_start=10,tiles='Stamen Toner')
#Draw circles
for x in centres.values:
    location = []
    location.append(x[4])
    location.append(x[5])
    map.circle_marker(location,radius=100,line_color='red',fill_color='red',popup="CRF: " + x[1])
```

```python
centre_ba = pd.read_csv("../data/croixrouge/position_centres_2015_banque_alimentaire.csv",sep=";")
# Banque alimentaires: point verts

for x in centre_ba.values:
    location = []
    location.append(x[1])
    location.append(x[2])
#    print(location)
    map.circle_marker(location,radius=100,line_color='green',fill_color='green',popup=x[0])

# restos du coeur : points bleus
centre_rc = pd.read_csv("../data/croixrouge/position_centres_2015_restos_du_coeur.csv",sep=";")
for x in centre_rc.values:
    location = []
    location.append(x[3])
    location.append(x[4])
#    print(location)
    map.circle_marker(location,radius=100,line_color='blue',fill_color='blue',popup="RC: " + x[0])
 
#inline_map(map)
```

```python
quartiers = pd.read_csv("../data/croixrouge/position_quartiers_prioritaires_2015.csv",sep="\t")

# Quartiers prioritaires: larges points noirs transparents
for x in quartiers.values:
    if x[1] > -10 and x[2] > -10:
        location = []
        location.append(x[2])
        location.append(x[1])
#        print(location)
        map.circle_marker(location,radius=1000,line_color='black',fill_color='black',popup=x[0],fill_opacity=0.2)
    
inline_map(map)
```
